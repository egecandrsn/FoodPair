```python
# Import necessary libraries
import pandas as pd
import ast
import re
from itertools import chain
from collections import Counter
import gzip
```

```python
# Upload data from drive to colab
df=pd.read_csv("/content/drive/MyDrive/foodpair.csv")
```

```python
# Turn columns to list of lists
df['lemmatized_ingredients'] = df['lemmatized_ingredients'].apply(eval)
```

```python
# It's a mess. Let's clean it up.
df.head()
```

```python
# Wow, look at this numbers. More than 2 million.
df.shape
```

```python
# There ara 178775 unique values.
ingredients = list(chain.from_iterable(df['lemmatized_ingredients']))
ingredient_counts = Counter(ingredients)
len(ingredient_counts.most_common())
```

```python
# Most common 30 ingredient.
ingredient_counts.most_common(30)
```

```python
#SEE VALUES
# With this code, we can check ingredients that contains "pasta ".
ingredient_regex = re.compile(r'.*pasta *', re.IGNORECASE)
ingredient_types = [(ingredient, count) for ingredient, count in ingredient_counts.most_common() if ingredient_regex.match(ingredient)]

for ingre, count in ingredient_types:
    print(f'{ingre}: {count}')
```

```python
# CHANGE ALL VALUES:
# To assign all pasta ingredients to just pasta, we can use code below.
df['lemmatized_ingredients'] = df['lemmatized_ingredients'].apply(
    lambda ingredients: ["pasta" if ingredient_regex.match(ingredient) else ingredient for ingredient in ingredients]
)
```

```python
# CHANGE ALL VALUES EXCEPT:
# But there is one ingredient that can be useful alone.
df['lemmatized_ingredients'] = df['lemmatized_ingredients'].apply(
    lambda ingredients: ["pasta" if (ingredient_regex.match(ingredient) and ingredient not in {"pasta sauce"}) else ingredient for ingredient in ingredients]
)
```

```python
# SINGLE VALUE REPLACE:
# Now we can assign pasta sauce to tomato sauce. But need to check tomato strings first and iterate over and over again.
df['lemmatized_ingredients'] = df['lemmatized_ingredients'].apply(
    lambda ingredients: ["pasta sauce" if ingredient == "tomato sauce" else ingredient for ingredient in ingredients]
)
```

```python
# SAVE CSV
# Don't forget to save all that hard-work.
df.to_csv('df.csv', index=False)
```

```python
df=pd.read_csv("df.csv",usecols=["lemmatized_ingredients"])
```

```python
my_text=df['lemmatized_ingredients']
df['lemmatized_ingredients'] = df['lemmatized_ingredients'].apply(ast.literal_eval)
my_text=df['lemmatized_ingredients'].to_list()
```

```python
split_point = len(my_text) // 2
part1 = my_text[:split_point]
part2 = my_text[split_point:]

# Create and save two Gzipped files
file_path1 = "part1.txt.gz"
file_path2 = "part2.txt.gz"

with gzip.open(file_path1, 'wb') as file1:
    for line in part1:
        file1.write(str(line).encode('utf-8') + b'\n')

with gzip.open(file_path2, 'wb') as file2:
    for line in part2:
        file2.write(str(line).encode('utf-8') + b'\n')
```
